### Segregating the data into X (independent) and y (target)

```python
y = df.iloc[:, (4)].values
```

```python
y
```

```python
# The columns 'Age' and 'AnnualSalary' are the features to use
X = df.iloc[:, [2,3]]
X
```

### Splitting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 0)
```

### Feature rescaling

```python
# let's perform rescaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

### Deploying K-Nearest Neighbour algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
model.fit(X_train, y_train)
```

```python
# predicting the test result
y_pred = model.predict(X_test)
y_pred
```

```python
# Creating the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (15,6))
sns.set(font_scale=1.1)
ax = sns.heatmap(cm, annot = True, fmt = 'd')
ax.set_xlabel("Prediction", fontsize=15, labelpad = 20)
ax.xaxis.set_ticklabels(["Positive", "Negative"])
ax.set_ylabel("Actual", fontsize=15, labelpad = 20)
ax.yaxis.set_ticklabels(['Positive', 'Negative'])
ax.set_title("Confusion Matrix for the car purchased Model", fontsize = 15, pad = 20)
plt.show()
```

The number of accurate prediction is 56+34 = 90 and the inaccurate prediction is 5+5 = 10. This shows that the performance of the model is good.

```python
mae = mean_absolute_error(y_pred, y_test)
print("The mean absolute error: ", mae)
```
